```python
example_data = pd.read_csv('../data/Recovery/recovery.csv').head()
example_data
```

```python
data = PandasData(example_data['article'], example_data['label'])
tokenized = BertBaseUncasedTokenizer().init().preprocess(data.copy())
```

```python
tokenized.features
```

```python
tokenized_list = tokenized.features.tolist()
tokenized_list
```

```python
encoder = BertBaseUncasedEncoder().init()
```

```python
tokenized_index=tokenized.features.index
```

```python
encodings = encoder.tokenizer(tokenized_list,
                                        is_split_into_words=True,
                                        padding=True,
                                        truncation=True,
                                        max_length=512)
encodings
```

```python
df_feature_encodings = pd.DataFrame(data=encodings.data, index=tokenized_index)
df_feature_encodings
```

```python
tokenized.features = df_feature_encodings
```
