### Encoding Tokens

```python
tokens = tokenizer.tokenize(sentence)
tokens
```

```python
tokenizer(tokens, is_split_into_words=True)
```

```python
feature_series = pd.Series([article[0:100], article[100:200], article[200:300]], name='article')
label_series = pd.Series([1, 0, 1], name='labels')
example_data = pd.concat([feature_series, labels_series], axis=1)
example_data = example_data.sample(frac=1, random_state=43)
example_data
```

```python
features = example_data['article']
tokenized_features = features.apply(lambda article: tokenizer.tokenize(article))
tokenized_features
```

```python
original_index = tokenized_features.index
original_index
```

```python
encoded = tokenizer(tokenized_features.tolist(), is_split_into_words=True, truncation=True, max_length=5)
encoded
```

```python
type(encoded)
```

```python
encoded_df = pd.DataFrame(data=encoded.data, index=original_index)
encoded_df
```

```python
data_df = pd.concat([encoded_df, labels_series], axis=1)
data_df
```
