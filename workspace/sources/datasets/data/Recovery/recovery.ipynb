```python
import pandas as pd
import matplotlib.pyplot as plt

from datasets.dataset import Dataset
from datasets.preprocessing.base import BertBasedUncasedTokenizer
from datasets.data.utils import plot_token_length_distribution


rd_seed = 42
```

### First Look

```python
df = pd.read_csv('recovery.csv')
df.info()
```

```python
df.head()
```

```python
label_counts = df['label'].value_counts()
plt.bar(label_counts.index, label_counts.values, color=['blue', 'orange'])
plt.xticks([0, 1], ['Label 0', 'Label 1'])
plt.title('Label Balance in Dataset')
plt.xlabel('Label')
plt.ylabel('Count')

# Add numbers on top of the bars
for i, value in enumerate(label_counts.values):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.show()
```
