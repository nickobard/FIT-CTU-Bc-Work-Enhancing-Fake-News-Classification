```python
from workspace.notebooks.utils import get_best_models_results
from workspace.notebooks.utils import render_template
import mlflow

mlflow.set_tracking_uri('../mlruns')
```

```python
# experiments = ['bert',
#                'distillibert',
#                'roberta',
#                'gpt1',
#                'gpt2']

experiments = ['bert',
               'distillibert', 'roberta']
datasets = ['recovery', 'isot', 'euvsdisinfo', 'cz_dataset']

experiments_to_notebooks_mapping = {'bert': 'bert_base_uncased',
                                    'gpt1': 'openai_gpt1',
                                    'gpt2': 'openai_gpt2'}
```

```python
for dataset_name in datasets:
    get_best_models_results(dataset_name,
                            experiments,
                            display=False)
    renderer_kwargs = {'templates_dir': 'tables_templates',
                       'experiment': None,
                       'dataset': dataset_name}
    render_template(template_name='best_models_table.tex.j2',
                    output_dir=f'assets/{dataset_name}/roc_curve/',
                    **renderer_kwargs)
```

```python
for experiment in experiments:
    notebook_dir = experiment
    if experiment in experiments_to_notebooks_mapping:
        notebook_dir = experiments_to_notebooks_mapping[experiment]
    %cd $notebook_dir
    %run results.ipynb
    %cd..
```
