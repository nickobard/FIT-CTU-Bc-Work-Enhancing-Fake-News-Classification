```python
from workspace.notebooks.utils import get_experiment_best_runs_by_metrics, \
    produce_experiment_model_hparams_tables, produce_experiment_pipeline_hparams_tables, \
    produce_experiment_confusion_matrices, produce_experiment_roc_curves, get_best_models_results, render_template

import mlflow

mlflow.set_tracking_uri('../../mlruns')
```

```python
experiment = 'bert'
```

```python
datasets = ['recovery', 'isot', 'euvsdisinfo', 'cz_dataset']

for dataset_name in datasets:
    best_runs = get_experiment_best_runs_by_metrics(experiment, dataset_name)

    renderer_kwargs = {'templates_dir': '../tables_templates/',
                       'experiment': experiment,
                       'dataset': dataset_name}

    produce_experiment_roc_curves(best_runs, dataset_name)
    render_template(template_name='roc_curves_table.tex.j2',
                    output_dir=f'assets/{dataset_name}/roc_curve/',
                    **renderer_kwargs)
    produce_experiment_confusion_matrices(best_runs, dataset_name)
    render_template(template_name='confusion_matrices_table.tex.j2',
                    output_dir=f'assets/{dataset_name}/confusion_matrix/',
                    **renderer_kwargs)
    produce_experiment_model_hparams_tables(best_runs, dataset_name)
    render_template(template_name='model_hparams_table.tex.j2',
                    output_dir=f'assets/{dataset_name}/model_hyperparameters/',
                    **renderer_kwargs)
    produce_experiment_pipeline_hparams_tables(best_runs, dataset_name)
    render_template(template_name='pipeline_hparams_table.tex.j2',
                    output_dir=f'assets/{dataset_name}/pipeline_hyperparameters/',
                    **renderer_kwargs)
```
